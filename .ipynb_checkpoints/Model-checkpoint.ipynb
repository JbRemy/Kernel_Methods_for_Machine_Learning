```python
os.chdir('.')
from Codes.dna2vec import dna2vec
from Codes.svm import svm
```

```python
D2V = dna2vec('Data/Xtr0.csv')
D2V.build_vocab(k_high=7, k_low=4, N=2000, vocab_size=4000)
print(D2V.vocab_size)
D2V.fit(epochs=5, window=5, n_negative=20, embeding_size=300, learning_rate=0.01)
```

```python
for i in range(3):
    with open('Data/Xtr{}.csv'.format(i), 'r') as file:
        S_list = file.readlines()

    for _ in range(len(S_list)):
        S_list[_] = S_list[_].strip()

    X = D2V.embed_data(S_list, 1000)
    #np.save('Data/X_{}_embeded'.format(i), X)
    print(i, "done")

for i in range(3):
    with open('Data/Xte{}.csv'.format(i), 'r') as file:
        S_list = file.readlines()

    for _ in range(len(S_list)):
        S_list[_] = S_list[_].strip()

    X = D2V.embed_data(S_list, 1000)
    #np.save('Data/X_{}_test_embeded'.format(i), X)
    print(i, "done")
```

```python
import pandas as pd
import time

start_time = time.time()

for i in range(3):
    y = np.array(pd.read_csv('Data/Ytr{}.csv'.format(i)))

    for _ in range(y.shape[0]):
        if y[_, 1] == 0:
            y[_, 1] = -1

    X = np.load('Data/X_{}_embeded.npy'.format(i))

    for reg in range(50, 500, 50):
        SVM = svm()
        SVM.fit(X[0:1600, :], y[0:1600, 1], reg, kernel='polynomial', l=1)
        y_pred_val, scores = SVM.predict(X[0:400, :])
        y_pred, scores = SVM.predict(X[1600:2000, :])
        err = 0
        err_val = 0
        for _ in range(400):
            if y_pred[_] != y[1600 + _, 1]:
                err += 1
            
            if y_pred_val[_] != y[_, 1]:
                err_val += 1
            
        print(i, reg, err/400, err_val/400, '%.2f' % (time.time()-start_time))
```
